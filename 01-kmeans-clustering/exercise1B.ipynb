## 5. Determine Optimal Number of Clusters `(10 pts)`

- Run K-Means for K = 1 to 10.
- Store WCSS (within-cluster sum of squares).
- Plot the elbow curve.

```python
# put your answer here
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_no)
    wcss.append(kmeans.inertia_)
```

```python
fig, axe = plt.subplots()
axe.plot(wcss)
axe.set_title('Elbow Method For Optimal K')
axe.set_xlabel('Number of Clusters (K)')
axe.set_ylabel('WCSS')
fig.show()
```

## 6. Train K-Means with Optimal K `(10 pts)`


- Choose K based on the elbow method.
- Fit the K-Means model.
- Assign cluster labels to the dataset.
- Display first 10 labeled rows.

```python
# put your answer here
cluster_kmeans = KMeans(n_clusters=3, random_state=42)
model_kmeans = cluster_kmeans.fit(X_no)
pred_kmeans = model_kmeans.labels_
print(pred_kmeans)
df['Cluster'] = pred_kmeans
display(df.head(10))
```
